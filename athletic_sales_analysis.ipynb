### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

            # Already provided example solution as
                #     Total_Products_Sold
                #     region	state	city	
                #     Northeast	New York	New York	111954
                #     South	Texas	Houston	90322
                #     West	California	San Francisco	85478
                #     Los Angeles	76384
                #     Southeast	Florida	Miami	73135

            # Basically asking to do this 
                # Group the data by region, state, and city and count the number of products sold
                    # products_sold = combined_sales.groupby(['region', 'state', 'city']).size().reset_index(name='products_sold')
                    # # Display the number of products sold for each region, state, and city
                    # print(products_sold)
                    # This code snippet will group the data by region, state, and city, count the number of products sold in each group, and then display the number of products sold for each region, state, and city.

products_sold = combined_sales.groupby(['region', 'state', 'city']).agg(Total_Products_Sold=('units_sold', 'sum'))
products_sold.rename(columns={'sum': 'Total_Products_Sold'}, inplace=True)
# Show the top 5 results.--- do not use products_sold.head(5)

        # Show the top 5 results
                     # Xpert learn
                            # top_5_results = products_sold_grouped.nlargest(5, 'Total_Products_Sold')
                            # print(top_5_results)
                            # This code will group the data by region, state, and city, count the number of products sold in each group, and then display the top 5 results showing the regions, states, and cities with the highest number of products sold.
top_5_results = products_sold.nlargest(5, 'Total_Products_Sold')
top_5_results
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.

                # Total_Products_Sold
                # region	state	city	
                # Northeast	New York	New York	111954
                # South	Texas	Houston	90322
                # West	California	San Francisco	85478
                # Los Angeles	76384
                # Southeast	Florida	Miami	73135

pivot_table_products_sold = combined_sales.pivot_table(index=['region', 'state', 'city'], values='units_sold', aggfunc='sum').reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table_products_sold.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
pivot_top_5_results = products_sold.nlargest(5, 'Total_Products_Sold')
pivot_top_5_results
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.

day_woman_ath_ftw = pd.pivot_table(woman_ath_ftw,values='total_sales', index='invoice_date', aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
day_woman_ath_ftw = day_woman_ath_ftw.rename(columns={"total_sales": "Total Sales"})

# Show the table.

day_woman_ath_ftw
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

            #Given as example results 
                # Total Sales
                # invoice_date	
                # 2021-07-16	1521825
                # 2021-12-16	1473497
                # 2021-06-17	1376988
                # 2021-08-17	1086294
                # 2021-07-23	1021806
                # 2021-11-17	1021145
                # 2021-12-09	915011
                # 2021-06-24	884238
                # 2021-07-09	869054
                # 2021-08-10	839120

        # Resample the pivot table into daily bins and calculate total sales for each day
                # daily_sales = pivot_table.resample('D').sum()
                # # Sort the resampled DataFrame in descending order of total sales
                # daily_sales_sorted = daily_sales.sort_values(by='total_sales', ascending=False)

                # # Display the resampled DataFrame with total sales for each day
                # print(daily_sales_sorted)
                # In this code:

                # pivot_table is the original pivot table with the total sales data.
                # resample('D') is used to resample the data into daily bins.
                # sum() calculates the total sales for each day.
                # sort_values() sorts the resampled DataFrame in descending order based on the total sales.
                # Finally, the resampled DataFrame daily_sales_sorted is displayed, showing the total sales for each day in descending order.

daily_sales = day_woman_ath_ftw.resample('D').sum()
# Sort the resampled pivot table in descending order on "Total Sales".

daily_sales_sorted = daily_sales.sort_values(by='Total Sales', ascending=False)
daily_sales_sorted
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.

weekly_sales = day_woman_ath_ftw.resample('W').sum()
# Sort the resampled pivot table in descending order on "Total Sales".
weekly_sales_sorted = weekly_sales.sort_values(by='Total Sales', ascending=False)
weekly_sales_sorted
```
